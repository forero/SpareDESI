```python
import numpy as np
from matplotlib import rcParams
import matplotlib.pyplot as plt
from astropy.table import Table
import json
```

```python
output_json = json.load(open('/global/cfs/cdirs/desi/users/forero/DR9mfiberassign/dark/qa.json'))
n_sky = list()
tile_id = list()
for k in output_json.keys():
    n_sky.append(output_json[k]['assign_sky'])
    tile_id.append(int(k))


tile_id = np.array(tile_id)
n_sky = np.array(n_sky)
n_avail = n_sky - 400
ii_tile = np.argsort(tile_id)

delta_pass = 2000
mean_n_avail = [np.mean(n_avail[(i//delta_pass)*delta_pass:(i//delta_pass)*delta_pass + delta_pass]) for i in range(0, len(n_sky))]

plt.figure(figsize=(8,6))
plt.plot(n_avail, label='PER TILE')
plt.plot(mean_n_avail, label='PER PASS', linewidth=5)
plt.xlabel('TILE')
plt.ylabel('N_AVAIL=N_SKY-400')
plt.legend()
plt.title("202101 Simulation 5 PASSES 10k tiles - TOTAL N_AVAIL fibers ={}\n ({:.1f} Million USED FIBERS IN TOTAL)".format(np.sum(n_sky-400), 5*len(n_sky)/1000))
```
